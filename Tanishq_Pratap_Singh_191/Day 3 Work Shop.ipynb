```python
Q5->

import numpy as np
from scipy import stats
data = np.array([12, 15, 14, 10, 13, 18, 20])
mean_value = stats.tmean(data)
print("Mean using scipy.stats:", mean_value)
t_stat, p_value = stats.ttest_1samp(data, 15)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
Q7->

from scipy import optimize
def func(x):
    return (x - 3)**2 + 4
result = optimize.minimize(func, x0=0)
print("Minimum Value:", result.fun)
print("Value of x at Minimum:", result.x)
```

```python
#Write a program to implement simpe linear regression

import numpy as np
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])
n = len(X)
mean_x = np.mean(X)
mean_y = np.mean(Y)
numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)
b1 = numerator / denominator
b0 = mean_y - b1 * mean_x
print("Intercept (b0):", b0)
print("Slope (b1):", b1)
Y_pred = b0 + b1 * X
print("Predicted values:", Y_pred)
plt.scatter(X, Y, color="blue", label="Actual Data")
plt.plot(X, Y_pred, color="red", label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Simple Linear Regression")
plt.show()
```
