```python
#1. Write a Python program to calculate the following for a given list of numbers:
import statistics
import numpy as np

numbers = [10, 20, 20, 30, 40, 50]

# Mean
mean = statistics.mean(numbers)

# Median
median = statistics.median(numbers)

# Mode
mode = statistics.mode(numbers)

# Variance
variance = statistics.variance(numbers)

# Standard Deviation
std_dev = statistics.stdev(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
#2. Create a NumPy array and perform the following operations:
import numpy as np

# Create NumPy array
arr = np.array([1, 2, 3, 4, 5, 6])

# Sum and Mean
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))

# Maximum and Minimum
print("Max:", np.max(arr))
print("Min:", np.min(arr))

# Reshape (2 rows, 3 columns)
reshaped = arr.reshape(2, 3)
print("Reshaped Array:\n", reshaped)

# Element-wise multiplication
result = arr * 2
print("Element-wise Multiplication:\n", result)
```

```python
#3. Write a Python program to:
import numpy as np

# Create two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Addition
print("Addition:\n", A + B)

# Subtraction
print("Subtraction:\n", A - B)

# Matrix Multiplication
print("Matrix Multiplication:\n", np.dot(A, B))

# Transpose
print("Transpose of A:\n", A.T)
```

```python
#5. Write a Python program using SciPy to:
from scipy import stats
import numpy as np

data = [10, 12, 9, 11, 10, 13, 12]

# Mean using scipy.stats
mean = stats.tmean(data)
print("Mean:", mean)

# One-sample t-test (compare mean with 10)
t_stat, p_value = stats.ttest_1samp(data, 10)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
#6. Write a Python program to compute definite integration of a mathematical function using
from scipy import integrate
import numpy as np

# Define function
def f(x):
    return x**2

# Integrate from 0 to 3
result, error = integrate.quad(f, 0, 3)

print("Definite Integration Result:", result)
print("Estimated Error:", error)
```

```python
#7. Write a Python program to minimize a mathematical function using scipy.optimize.
from scipy.optimize import minimize

# Define function
def f(x):
    return x**2 + 4*x + 4

# Initial guess
x0 = 0

# Minimize function
result = minimize(f, x0)

print("Minimum Value:", result.fun)
print("At x =", result.x)
```

```python
#classwork-5: Linear Regression
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 5, 4, 5])

# Model
model = LinearRegression()
model.fit(x, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
#Classwaork-6: Linear Regression using CSV File
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load CSV file
file_path = '/content/data1.csv'  # change this to your file path
data = pd.read_csv(file_path)

# Step 2: Check the data
print(data.head())

# Step 3: Select independent and dependent variables
# Example: X = 'age', Y = 'sales' (change according to your CSV column names)
X = data['age'].values.reshape(-1, 1)   # independent variable
Y = data['sales'].values               # dependent variable

# Step 4: Create Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Step 5: Get slope and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)

# Step 6: Predict values
Y_pred = model.predict(X)

# Step 7: Plot the regression line
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('age')     # change to your X column name
plt.ylabel('sales')  # change to your Y column name
plt.legend()
plt.show()
```
