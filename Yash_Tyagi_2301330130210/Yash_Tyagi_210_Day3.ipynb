```python
# 1. Write a Python program to calculate the following for a given list of numbers:
import statistics
import math

# Given list
numbers = [10, 20, 30, 40, 50, 20]

# Mean
mean_value = statistics.mean(numbers)

# Median
median_value = statistics.median(numbers)

# Mode
mode_value = statistics.mode(numbers)

# Variance
variance_value = statistics.variance(numbers)

# Standard Deviation
std_deviation = statistics.stdev(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_deviation)
```

```python

# 5. Write a Python program using SciPy to:
import numpy as np
from scipy import stats

# Sample data
data = [10, 12, 9, 11, 10, 13, 12]

# Mean using scipy.stats
mean_value = stats.tmean(data)
print("Mean using scipy.stats:", mean_value)

# Perform one-sample t-test (compare against population mean = 10)
t_stat, p_value = stats.ttest_1samp(data, 10)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
# 6. Write a Python program to compute definite integration of a mathematical function using scipy.integrate.
from scipy import integrate
import numpy as np

# Define function
def func(x):
    return x**2

# Integrate from 0 to 3
result, error = integrate.quad(func, 0, 3)

print("Definite Integration Result:", result)
print("Estimated Error:", error)
```

```python
# 7. Write a Python program to minimize a mathematical function using scipy.optimize.
from scipy import optimize

# Define function
def func(x):
    return x**2 + 5*x + 6

# Minimize function
result = optimize.minimize(func, x0=0)

print("Minimum Value:", result.fun)
print("Value of x at Minimum:", result.x)
```

```python
# Program to implement Simple Linear Regression

def linear_regression(x, y):
    n = len(x)

    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum([x[i] * y[i] for i in range(n)])
    sum_x2 = sum([x[i] ** 2 for i in range(n)])

    # Calculating slope (b1) and intercept (b0)
    b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
    b0 = (sum_y - b1 * sum_x) / n

    return b0, b1


# Example data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

b0, b1 = linear_regression(x, y)

print("Intercept (b0):", b0)
print("Slope (b1):", b1)

# Predicting a value
x_new = 6
y_pred = b0 + b1 * x_new
print("Predicted value at x =", x_new, "is", y_pred)
```
